Create Dummy Variables:

-pd.get_dummies() is a pandas function that converts categorical variables into dummy/indicator variables.
-Each unique category value in a categorical column is converted into a new binary column.
-drop_first=True indicates that the first category of each categorical variable will be dropped to avoid the dummy variable trap.

Drop Original Variable for Which the Dummy Was Created:

When pd.get_dummies() is used, it automatically drops the original categorical columns and replaces them with the new dummy variables.
Drop First Dummy Variable for Each Set of Dummies Created:

drop_first=True helps to avoid multicollinearity in regression models by dropping the first dummy variable in each set.
This ensures that the dummy variables are linearly independent, which is important for some machine learning models like linear regression.

**MODEL 1:**

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Convert boolean columns to integers
X_train_rfe = X_train_rfe.applymap(lambda x: int(x) if isinstance(x, bool) else x)

# Ensure there are no non-numeric columns
X_train_rfe = X_train_rfe.select_dtypes(include=[np.number])

# Ensure there are no missing values
X_train_rfe = X_train_rfe.fillna(X_train_rfe.mean())

# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by="VIF", ascending=False)

print(vif)
```

```python
import statsmodels.api as sm

# add a constant
X_train_lm1 = sm.add_constant(X_train_rfe)

# create a first fitted model
lr1 = sm.OLS(y_train,X_train_lm1.astype(float)).fit()
```

```python
# Check the parameters obtained

lr1.params
```

```python
# Print a summary of the linear regression model obtained
print(lr1.summary())
```
